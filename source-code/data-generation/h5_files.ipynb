## Requirements

```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

## Row-major storage

```python
h5_file = h5py.File('data/imgs_row_major.h5')
```

```python
h5_file['data'].shape
```

```python
plt.imshow(h5_file['data'][0]);
```

```python
plt.imshow(h5_file['data'][-1]);
```

```python
h5_file.close()
```

## Column-major storage

```python
h5_file = h5py.File('data/imgs_col_major.h5')
```

```python
h5_file['data'].shape
```

```python
plt.imshow(h5_file['data'][:, :, :, 0]);
```

```python
plt.imshow(h5_file['data'][:, :, :, -1]);
```

```python
h5_file.close()
```
